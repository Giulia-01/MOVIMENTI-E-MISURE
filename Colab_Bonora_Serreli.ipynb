##  **Processamento**

Si procede con la pulizia dei dati con le seguenti operazioni:

## **RQ1 — Quali generi sono più comuni nelle varie epoche?**

---

Si è scelto il **secolo** come unità di epoca.

Operazioni:
1. Si crea un nuovo DataFrame chiamato `df_generi `che contenga solo le colonne *secolo* e *generi* del df originale e si svincola tramite la funzione `.copy()`, così da evitare modifiche al df originale.

2. Quindi si crea una nuova colonna in `df_generi` chiamata `genere_singolo`. Tramite la funzione `.str.split(;)` si crea una lista di stringhe prelevandole dalla colonna 'generi', i cui elementi sono separati dal ';'.

2. Si procede eliminando spazi extra (`.str.strip()`) e mettendo in minuscolo (`.str.lower()`) i valori contenuti nella colonna genere_singolo.

3. Si eliminano eventuali valori nulli e vuoti.  

4. Si può così contare quanti generi ci sono per ogni secolo.

5. Visualizziamo i top 5 di ogni secolo tramite le funzioni `.sort_values` e `.groupby`



```python
df_generi = df[["secolo", "generi"]].copy()
df_generi["genere_singolo"] = df_generi["generi"].str.split(";")
df_generi = df_generi.explode("genere_singolo")

df_generi["genere_singolo"] = df_generi["genere_singolo"].str.strip().str.lower()

df_generi = df_generi.dropna(subset=["secolo", "genere_singolo"])
df_generi = df_generi[df_generi["genere_singolo"] != ""]

conteggi_secolo = df_generi.groupby(["secolo", "genere_singolo"]).size().reset_index(name="quantita")

top5_per_secolo = conteggi_secolo.sort_values(["secolo", "quantita"], ascending=[True, False])
top5_per_secolo = top5_per_secolo.groupby("secolo").head(5)

top5_per_secolo
```

## **RQ1 (seconda parte) — Esiste un legame tra movimenti e generi?**
---
Esempio:

- nel movimento etichettato **Romanticismo** (o “pittura romantica”) quali sono i  **generi** che compaiono più spesso?
- successivamente si confronta con la popolarità di quel genere in altri movimenti.

**Operazioni**

1. Si procede estraendo dal DataFrame df i dati della colonna `movimenti` e, laddove presenti stringhe col carattere `;`, si applica `.str.split()`.

2. Tramite `.explode()` si attribuisce ad ogni valore una propria riga. con la funzione .unique() si assicura che non vi siano ripetizioni.

3. Dopo aver escluso i valori nulli e stringhe vuote tramite `pd.notna()` e `!= ''`, si può procedere alla stampa.



```python
colonna_movimenti = df['movimenti'].str.split(';').explode().str.strip().str.lower().unique()

print("Tutti i movimenti presenti nel dataset:")
for movimento in colonna_movimenti:
    if pd.notna(movimento) and movimento != '':
        print(f"- {movimento}")
```

Analizzando l'elenco dei movimenti, si è optato per utilizzare tre macrocategorie per raggruppare movimenti simili o rinconducibili allo stesso periodo storico-artistico: **Rinascimento**, **Barocco** e **Romanticismo**.

1. Si filtra nella colonna `movimenti` tutto ciò che contiene la parola *rinascim-* e gli si assegna la variabile `df_rinascimento`.

2. Si procede con la pulizia dei dati della colonna `generi`.

3. Infine, dopo aver ottenuto il conteggio per ciascun genere, si visualizza la riuscita dell'operazione con `.head(5)`.


Queste operazioni vengono ripetute per analizzare i generi di Barocco e Romanticismo.

```python
df_rinascimento = df[df['movimenti'].str.contains('rinascim', na=False, case=False)]
generi_rinascimentali = df_rinascimento['generi'].str.split(';').explode().str.strip().str.lower()
top_rinascimento = generi_rinascimentali[generi_rinascimentali != ""].value_counts().head(5)

df_barocco = df[df['movimenti'].str.contains('barocc', na=False, case=False)]
generi_barocchi = df_barocco['generi'].str.split(';').explode().str.strip().str.lower()
top_barocco = generi_barocchi[generi_barocchi != ""].value_counts().head(5)

df_romanticismo = df[df['movimenti'].str.contains('romant', na=False, case=False)]
generi_romantici = df_romanticismo['generi'].str.split(';').explode().str.strip().str.lower()
top_romanticismo = generi_romantici[generi_romantici != ""].value_counts().head(5)

print("TOP GENERI RINASCIMENTO:")
print(top_rinascimento)
print("\nTOP GENERI BAROCCO:")
print(top_barocco)
print("\nTOP GENERI ROMANTICISMO:")
print(top_romanticismo)
```

## **RQ2 — Esiste una relazione tra dimensione dell'opera (altezza × larghezza) e genere?**

---
**Operazioni**

1. Calcolo del’**area** in metri quadrati = altezza × larghezza (quando entrambe presenti). Per convertire le misure da cm a m si divide per 100.

2. Si separano i generi con `.str.split()` e `.explode()`.

3. Usando `.mean()`, si esegue il calcolo della media per ogni genere.

4. Per fare un confronto, si tengono solo i generi che hanno almeno 30 opere (per avere dati affidabili).




```python
df_misure = df.dropna(subset=["altezza", "larghezza"]).copy()
df_misure["area_m2"] = (df_misure["altezza"] / 100) * (df_misure["larghezza"] / 100)

df_misure_generi = df_misure[["area_m2", "generi"]].copy()
df_misure_generi["genere"] = df_misure_generi["generi"].str.split(";")
df_misure_generi = df_misure_generi.explode("genere")
df_misure_generi["genere"] = df_misure_generi["genere"].str.strip().str.lower()
df_misure_generi = df_misure_generi[df_misure_generi["genere"] != ""]

statistiche_area = df_misure_generi.groupby("genere")["area_m2"].mean().sort_values(ascending=False)

conteggio_opere = df_misure_generi["genere"].value_counts()
generi_con_abbastanza_dati = conteggio_opere[conteggio_opere >= 30].index

risultato_finale = statistiche_area[statistiche_area.index.isin(generi_con_abbastanza_dati)]

print(risultato_finale.head(10))
```

## **Confronto dei risultati della media delle aree per genere**

**Arte religiosa**: Con una superficie media di circa 10.94 m², rimane il genere con le opere di dimensioni maggiori. Questo suggerisce che, in media, le opere religiose tendono ad essere di grande formato.

**Pittura mitologica**: Segue con una media di circa 3.38 m², indicando che anche per questo genere la predilezione di dimensioni considerevoli, sebbene inferiori all'arte religiosa.

**Ritratto**: La media scende a circa 2.27 m², confermando che i ritratti hanno dimensioni mediamente più contenute rispetto ai grandi quadri religiosi o mitologici.

**Arte di genere**: Si attesta intorno a 1.27 m².

**Autoritratto**: Con una media di circa 0.52 m², è il genere che presenta le dimensioni mediamente più piccole tra quelli elencati, il che è atteso data la natura più intima di questo genere.

---

In sintesi, nel dataset esiste una relazione tra dimensioni (m²) e genere: religioso/mitologico tende a formati più grandi, mentre ritratto/autoritratto tende a formati più piccoli.
